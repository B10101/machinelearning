```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# sample data

stamps_bought = np.array([1,3,5,7,9]).reshape((-1,1))
amount_spent = np.array([2,6,8,12,18])

```

```python
# linear regression model

model = LinearRegression()

model.fit(stamps_bought, amount_spent)


#predict

next = 10
predict = model.predict([[next]])
predict
```

```python
#plot

plt.scatter(stamps_bought, amount_spent, color = 'red')
plt.plot(stamps_bought, model.predict(stamps_bought), color = 'green')
plt.title("Prediction")
plt.xlabel("stamps")
plt.ylabel("amount")
```
